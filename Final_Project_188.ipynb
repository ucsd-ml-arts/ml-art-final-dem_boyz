```
import os
import cv2
import numpy as np
import math
from scipy.io import wavfile as wav
from scipy.fftpack import fft
from scipy.signal import spectrogram # x: data before reshape; fs: 44100 smp fq
# nperseg
from sklearn.preprocessing import normalize
import numpy as np

VIDEOS_PATH = './'
VIDEOS_EXTENSION = '.mkv'  # for example
AUDIO_EXT = 'wav'

EXTRACT_VIDEO_COMMAND = ('ffmpeg -i "{from_video_path}" '
                         '-f {audio_ext} -ab 192000 '
                         '-vn "{to_audio_path}"')

index = 1
oof = 10 #IMPORTANT
os.chdir(VIDEOS_PATH)
files = os.listdir(VIDEOS_PATH)
for f in files:
    if 'Nines' not in f:
        continue
    if not f.endswith(VIDEOS_EXTENSION):
      print('*')
      continue

    audio_file_name = '{}.{}'.format(f, AUDIO_EXT)
    print(audio_file_name)
    command = EXTRACT_VIDEO_COMMAND.format(
        from_video_path=f, audio_ext=AUDIO_EXT, to_audio_path=audio_file_name,
    )
    os.system(command)
    #read video
    cap = cv2.VideoCapture(str(f))
    frameCount = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    frameWidth = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frameHeight = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    
    buf = np.empty((frameCount, frameHeight, frameWidth, 3), np.dtype('uint8'))

    fc = 0
    ret = True
    fc = 1
    frameCount = 90*30
    while (fc < frameCount  and ret):
      if (fc < 30*30):
        cap.read()
      else:  
        ret, buf[fc] = cap.read()
      fc += 1

    cap.release()
    print(0)
    cv2.waitKey(0)


    #split into 1.5 sec chunks
    #buf = buf[30*30:90*30:45]
    buf = buf[(30+oof)*30:(31+oof)*30:1]
    #resize
    siz = np.shape(buf)
    arr1 = np.linspace(0,siz[1]-1,256)
    arr2 = np.linspace(0,siz[2]-1,256)
    x1 = [];
    x2 = [];

    for x in arr1:
      x1.append(math.floor(x))
    
    for x in arr2:
      x2.append(math.floor(x))
      
    buf = buf[:,x1,:,:]
    buf = buf[:,:,x2,:]
                           
    rate, data = wav.read(audio_file_name)
    a = np.asarray(data)
    data = a[:,0]
    num_imgs = int(np.floor(len(data)/65536))
    print(1)
    norm_windows = []
    for j in range((29+oof)*30, (30+oof)*30):
      window = np.zeros(65536)
      for i in range(0, 65536):
        window[i] = data[i + j*1470]
      #print(window.shape)
      f, t, Sxx = spectrogram(window, 44100, nperseg= 288, noverlap = 33, nfft = 510)
      #print(Sxx.shape)
      #window = spectrogram(window, 44100)
      #window_reshaped = np.reshape(window, (256, 256))
      norm_windows.append(Sxx)
    print(2)
             
    images = np.stack((norm_windows,norm_windows,norm_windows),axis = 3)
    
    #combine image and sound
    full_im = np.concatenate((buf,images[0:len(buf)]),axis = 2)
    print(3)
    index = index + 30*oof
    #write to file
    for el in full_im:
      string = "./tests/" + str(index) + ".jpg"
      print(string)
      cv2.imwrite(string,el)
      index = index + 1
      
    del(buf)
    del(images)
    del(full_im)
    del(norm_windows)
```

```
from scipy.io import wavfile as wav
from scipy.fftpack import fft
import numpy as np
rate, data = wav.read('Kanye West - Fade (Explicit).mp4.wav')
fft_out = fft(data)
```

```
a = np.asarray(data)
print(np.shape(a))
```

```
data = a[:,0]
num_imgs = int(np.floor(len(data)/65536))
print(num_imgs)
```

```
windows = []
for j in range(0, num_imgs):
  window = np.zeros(65536)
  for i in range(0, 65536):
    window[i] = data[i + j*65536]
  window_reshaped = np.reshape(window, (256, 256))
  windows.append(window_reshaped)

print(windows)
```

```
norm_windows = []
for window in windows:
  max_val = np.amax(abs(window))
  window = window/max_val
  norm_windows.append(window)

print(norm_windows)
```
